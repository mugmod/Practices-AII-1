# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def list_comp(a,b):
  res = []

  for num1 in a:
    b.append(num1)

  for num2 in b:
    if b.count(num2) < 2:
      res.append(num2)

  return res


a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

print(f'Уникальные числа/число : {list_comp(a,b)}')
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import random


def adoption(pet_personality, pets):
  print('Ваш питомец:', *random.choices(pet_gender))

  name = input('Введите имя для своего питомца: ')

  if name in pets:
    print(f'''Питомец с таким именем уже существует, придумайте другое имя,
используйте фантазию''')
  else:
    age = int(input('Введите возраст своего питомца в месяцах (просто цифру) : '))
    pet_personality['age'] = age
    pets[name] = pet_personality

    print('Имя и возраст добавлены')


def hunger_end(pets, name):
  print(f'Голод вашего питомца: {pets[name]["hunger"]}')
  print('Ваш чадо съел пачку мяса слона')

  pets[name]["hunger"] += 20

  print(f'Теперь шкала голода вашего питомца составляет: {pets[name]["hunger"]}')


def play(pets, name):
  print(f'Ваш питомец весел с коэффециентом: {pets[name]["joy"]}')
  print('''Можно кинуть питомцу палку - 1
Можно кинуть в питомца палку - 2
''')

  command = int(input("Как поиграть с питомцем? 1 или 2: "))

  if command == 1:
    pets[name]["joy"] += 20

    print(f'Вы кинули питомцу палку, его веселье достигло: {pets[name]["joy"]}')
  elif command == 2:
    pets[name]["joy"] += 20

    print(f'Вы так себе хозяин, но веселье питомца достигло: {pets[name]["joy"]}')


def sleep(pets, name):
  print(f'Коэффециент энергии вашего питомца равен: {pets[name]["energy"]}')

  pets[name]["energy"] += 20

  print(f'Вы поспали ровно ночь, теперь энергия вашего питомца равна: {pets[name]["energy"]}')


def personality(pets, name):
  print(f''' Имя питомца: {name}
Уровень голода: {pets[name]["hunger"]}
Уровень веселья: {pets[name]["joy"]}
Уровень энергии: {pets[name["energy"]]}
''')


pet_gender = ['кошка', 'собака', 'черепаха']
pet_personality = {'hunger': 20, 'joy': 20, 'energy': 20}
pets = {}
command = None

while command != 6:
  print('''
                  Тамогочик
Вот то, что вы можете делать:
1 - усыновить питомца
2 - покормить питомца
3 - поиграть с питомцем
4 - уложить спать питомца
5 - посмотреть состояние питомца
6 - уйти от ответственности
''')

  name = ''
  command = int(input("Что выберете? Введите цифру "))

  if command == 1:
    print('''
    Ура!!! Вы усыновили питомца!
Это конечно не семейная ипотека, но тоже круто
''')

    adoption(pet_personality, pets)

  elif command == 2:
    print('''Вы выбрали покормить питомца
''')

    if name == '':
      name = input("Введите имя питомца ")

      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)

        hunger_end(pets, name)
        pets[name]['joy'] -= 5
        pets[name]['energy'] -= 5
      else:
        print("У вас нет питомца с таким именем")

    elif name != '':
      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)

        hunger_end(pets, name)
        pets[name]['joy'] -= 5
        pets[name]['energy'] -= 5

    else:
      print('''
У вас нет питомца, усыновите нового, или питомца с таким именем нет
''')

  if command == 3:
    print('''
Вы решили поиграть с питомцем
''')

    if name == '':
      name = input("Введите имя питомца ")

      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)

        play(pets, name)
        pets[name]['hunger'] -= 5
        pets[name]['energy'] -= 5
      else:
        print("У вас нет питомца с таким именем")

    elif name != '':
      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)
        play(pets, name)
        pets[name]['hunger'] -= 5
        pets[name]['energy'] -= 5

    else:
      print('''
У вас нет питомца, усыновите нового, или питомца с таким именем нет
''')

  if command == 4:
    print('''
Вы решили поспать с питомцем
''')

    if name == '':
      name = input("Введите имя питомца ")

      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)

        sleep(pets, name)
        pets[name]['hunger'] -= 5
        pets[name]['joy'] -= 5
      else:
        print("У вас нет питомца с таким именем")

    elif name != '':
      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)

        sleep(pets, name)
        pets[name]['hunger'] -= 5
        pets[name]['joy'] -= 5

    else:
      print('''
У вас нет питомца, усыновите нового, или питомца с таким именем нет
''')

  if command == 5:
    print('''
Вы решили посмотреть уровни характеристик питомца
''')

    if name == '':
      name = input("Введите имя питомца ")

      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)

        personality(pets, name)
      else:
        print("У вас нет питомца с таким именем")

    elif name != '':
      if name in pets:
        for x in pets[name]:
          if pets[name][x] < 1:
            print("Ваш питомец умер")

            pets.pop(name, None)

        personality(pets, name)

    else:
      print('''
У вас нет питомца, усыновите нового, или питомца с таким именем нет
''')

  if command == 6:
    print('''
Все питомцы успешно сбежали через открытое окно и больше к вам не вернулись
''')
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random


class Personage:
    def __init__(self, name):
        armor = ['wooden armor', 'iron armor', 'naked body']
        weapon = ['stick', 'machine', 'gun', 'sword']
        damage = [50, 70, 45, 25, 63, 32]

        self.name = name
        self.armor = random.choice(armor)
        self.weapon = random.choice(weapon)
        self.damage = random.choice(damage)
        self.health = 100


    def get_health(self):
        return self.health


    def get_damage(self):
        return self.damage


    def take_damage(self, damage):
        self.health -= damage

        if self.health > 0:
            return f'Урон нанесён со стороны дракона, хп рыцаря: {self.health}'
        else:
            return 'Последний урон нанесён, хп рыцаря на нуле, рыцарь умер'


class Dragon:
    def __init__(self, name):
        health = [100, 150, 200, 250, 300, 350, 400, 450, 550, 600]
        damage = [20, 13, 4, 5, 6, 12]

        self.name = name
        self.damage = random.choice(damage)
        self.health = random.choice(health)


    def get_health(self):
        return self.health


    def get_damage(self):
        return self.damage


    def take_damage(self, damage):
        self.health -= damage

        if self.health > 0:
            return f'Урон нанесён со стороны рыцаря, хп дракона: {self.health}'
        else:
            return 'Последний урон нанесён, хп дракона на нуле, дракон умер'


personage_name = input("Введите имя для персонажа: ")
dragon_name = input("Введите имя для дракона: ")

personage = Personage(personage_name)
dragon = Dragon(dragon_name)

while personage.get_health() > 0 and dragon.get_health() > 0:
    if random.choice([True, False]):
        print(dragon.take_damage(personage.get_damage()))
    else:
        print(personage.take_damage(dragon.get_damage()))

if personage.get_health() <= 0:
    print(f'{personage.name} умер, {dragon.name} победил!')
else:
    print(f'{dragon.name} умер, {personage.name} победил!')
````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests as req


def get_pokemon(pokemons):
    name = input('Введите имя покемона (с маленькой буквы), которого хотите добавить: ')
    response = req.get(f'https://pokeapi.co/api/v2/pokemon/{name}')

    if response.status_code == 200:
        data = response.json()

        name = data['name']
        weight = data['weight']
        height = data['height']
        tipe = data['types'][0]['type']['name']
        message1 = f'{name} - Тип: {tipe}. Вес: {weight}. Рост: {height}.'
        abilities = []

        for i in data['abilities']:
            abilities.append(i['ability']['name'])

        message2 = 'Способности:', *abilities

        if name not in pokemons:
            pokemons[name] = {'info': {}}

        pokemons[name]['info']['ab1'] = message1
        pokemons[name]['info']['ab2'] = message2

        print('Покемон добавлен')
    else:
        print(f'Ошибка сервера, возможно, что такого покемона нет: {response.status_code}')


def pokemon_delete(pokemons):
    name = input('Введите имя покемона: ')

    if name in pokemons:
        del pokemons[name]

        print('Покемон исключён из команды')
    else:
        print('Такого покемона нет в команде')


def pokemon_search(pokemons):
    name = input("Ввдите имя покемона: ")

    if name in pokemons:
        print(f'''
        Имя покемона: {name}
        Информация: {pokemons[name]["info"]}
              ''')
    else:
        print('Такого покемона нет в команде')


def pokemons_info(pokemons):
    for pok in pokemons:
        print(f'''Покемон: {pok}
        Информация: {pokemons[pok]['info']['ab1']}
                    {pokemons[pok]['info']['ab2']}
        ''')


def pokemon_battle(pokemons):
    name1 = input('Введите имя первого покемона: ')
    name2 = input('Введите имя второго покемона: ')

    if name1 not in pokemons:
        print(f'Покемон {name1} не найден в команде.')
        return
    if name2 not in pokemons:
        print(f'Покемон {name2} не найден в команде.')
        return

    weight1 = int(pokemons[name1]['info']['ab1'].split('Вес: ')[1].split('.')[0])
    height1 = int(pokemons[name1]['info']['ab1'].split('Рост: ')[1].split('.')[0])

    weight2 = int(pokemons[name2]['info']['ab1'].split('Вес: ')[1].split('.')[0])
    height2 = int(pokemons[name2]['info']['ab1'].split('Рост: ')[1].split('.')[0])

    score1 = weight1 + height1
    score2 = weight2 + height2

    if score1 > score2:
        print(f'Победил {name1}!')
    elif score2 > score1:
        print(f'Победил {name2}!')
    else:
        print('Ничья!')


pokemons = {}
command = None
while command != 0:
    print('''
    Функции бота:
    1 - Добавить покемона в команду
    2 - Удалить покемона из команды
    3 - Посмотреть информацию о каждом покемоне в команде
    4 - Найти покемона по имени (вывод есть ли покемон
    в команде и информацию о нём)
    5 - Устроить тренировочный бой
    0 - Завершить программу
    ''')

    command = int(input('Введите цифру: '))

    if command == 0:
        print('''
        Завершить программу''')
    elif command == 1:
        print('''
        Добавить покемона в команду
        ''')

        get_pokemon(pokemons)
    elif command == 2:
        print('''
        Удалить покемона из команды
        ''')

        pokemon_delete(pokemons)
    elif command == 3:
        print('''
        Посмотреть информацию о каждом покемоне в команде
        ''')

        pokemons_info(pokemons)
    elif command == 4:
        print('''
        Найти покемона по имени (вывод есть ли покемон
        в команде и информацию о нём)
        ''')

        pokemon_search(pokemons)
    elif command == 5:
        print('''
        Устроить тренировочный бой
        ''')

        pokemon_battle(pokemons)
    else:
        print('''
        Введите число из списка функций
        ''')

        continue
````
